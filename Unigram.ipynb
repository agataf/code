```julia
### modify Plots.jl function so that many tick labels can be shown
```

```julia
using Plots
gr(size=(2000,500))
# Plots.scalefontsizes(2)  # You may need to uncomment/modify this statement if fonts are too small

# modify Plots.jl function so that all ticks are shown
function Plots.get_ticks(axis::Plots.Axis)
    ticks = axis[:ticks]
    ticks in (nothing, false) && return nothing

    dvals = axis[:discrete_values]
    cv, dv = if !isempty(dvals) && ticks == :auto
        # discrete ticks...
        axis[:continuous_values], dvals
    elseif ticks == :auto
        # compute optimal ticks and labels
        Plots.optimal_ticks_and_labels(axis)
    elseif typeof(ticks) <: Plots.AVec
        # override ticks, but get the labels
        Plots.optimal_ticks_and_labels(axis, ticks)
    elseif typeof(ticks) <: NTuple{2}
        # assuming we're passed (ticks, labels)
        ticks
    else
        error("Unknown ticks type in get_ticks: $(typeof(ticks))")
    end
    # @show ticks dvals cv dv

    # MODIFICATION: increased cutoff value for sampling ticks
    if length(cv) > 100
        rng = Int[round(Int,i) for i in linspace(1, length(cv), 15)]
        cv[rng], dv[rng]
    else
        cv, dv
    end
end
```

### read in the corpus from a text file

```julia
fid = open("shakespeare.txt")
corpus = readstring(fid)
close(fid)
```

### compute unigram statistics (counts)

```julia
alphabet = unique(corpus)
alphabetlength = length(alphabet)
counts = zeros(alphabetlength)
for i in 1:alphabetlength
    counts[i] = count(c -> c == alphabet[i], corpus)
end
```

### sort and plot histogram with characters ordered so counts decrease

```julia
p = sortperm(counts, rev=true)
bar(counts[p], xticks = (vec(1:alphabetlength), alphabet[p]), legend=:none)
```
